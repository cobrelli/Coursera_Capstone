```python
# Read the geospatial data into pandas
toronto_coordinates_per_postal_code = pd.read_csv('Geospatial_Coordinates.csv')
```

```python
toronto_coordinates_per_postal_code
```

Next we will merge the dataframes into one dataframe using the postal code as the index and then drop the extra 'Postal Code' column.

```python
toronto_neighbourhoods_geospatial = toronto_neighbourhoods.merge(right=toronto_coordinates_per_postal_code, how='left', left_on='PostalCode', right_on='Postal Code')
toronto_neighbourhoods_geospatial.drop('Postal Code', axis=1, inplace=True)
```
